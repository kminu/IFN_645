# Clustering 1 - Based on the time they spend

```python
#Drop item?

data1 = get_data()
data1.drop(['OperatingSystems'],axis=1, inplace=True)
data1.drop(['Browser'],axis=1, inplace=True)
data1.drop(['Region'],axis=1, inplace=True)
data1.drop(['Administrative'],axis=1, inplace=True)
data1.drop(['Informational'],axis=1, inplace=True)
data1.drop(['ProductRelated'],axis=1, inplace=True)
data1.drop(['Month'],axis=1, inplace=True)
data1.drop(['SpecialDay'],axis=1, inplace=True)
data1.drop(['TrafficType'],axis=1, inplace=True)
data1.drop(['Weekend'],axis=1, inplace=True)
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# random state, we will use 42 instead of 10 for a change
rs = 42

# set the random state. different random state seeds might result in different centroids locations
model = KMeans(n_clusters=3, random_state=rs)
model.fit(data1)

# sum of intra-cluster distances
print("Sum of intra-cluster distance:", model.inertia_)

print("Centroid locations:")
for centroid in model.cluster_centers_:
    print(centroid)

# assign cluster ID to each record in X
# Ignore the warning, does not apply to our case here
y = model.predict(data1)
data1['Cluster_ID'] = y

# how many records are in each cluster
print("Cluster membership")
print(data1['Cluster_ID'].value_counts())

# pairplot the cluster distribution.
cluster_g = sns.pairplot(data1, hue='Cluster_ID')
plt.show()
```

```python
# list to save the clusters and cost
clusters = []
inertia_vals = []

# this whole process should take a while
for k in range(2, 15, 2):
    # train clustering with the specified K
    model = KMeans(n_clusters=k, random_state=rs, n_jobs=10)
    model.fit(data1)
    
    # append model to cluster list
    clusters.append(model)
    inertia_vals.append(model.inertia_)
```

```python
# plot the inertia vs K values
plt.plot(range(2,15,2), inertia_vals, marker='*')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

print(clusters[1])
print("Silhouette score for k=4", silhouette_score(data1, clusters[1].predict(data1)))

print(clusters[2])
print("Silhouette score for k=6", silhouette_score(data1, clusters[2].predict(data1)))

print(clusters[3])
print("Silhouette score for k=8", silhouette_score(data1, clusters[3].predict(data1)))
```
